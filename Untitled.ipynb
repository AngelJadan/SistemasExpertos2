# Para varios datos

#### 1. Creamos 2 arreglos con las probabilidades p(a) y probabilidad b dado a p(b|a)

```python
pa=[0.4, 0.35, 0.25 ]
pba=[0.15, 0.20, 0.65]
```

#### 2. Declaramos variables y los tamaos de cada arreglo.

```python
c=0
s=0
tpa=len(pa)
tpba=len(pba)
i=0
j=0

s=0
```

#### 3. Realizamos la suma de las probabilidades dado a

```python
while i <tpba:
    s=s+(pa[i]*pba[i])
    i=i+1
```

#### 4. Realizamos Probabilidad de cada uno.

```python
while j<tpa:
    pr=(pa[j]*pba[j])/s
    print("Pr"+str(j)+": "+str(pr))
    j=j+1
```
